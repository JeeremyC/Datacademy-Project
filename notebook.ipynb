```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('/work/largest_us_retailers.csv')
df
```

## I. Business questions

### 1. What is the average of sales? (not including the dominant company)

```
average = round(df['Sales'][1:].mean(), 2)
print(f'The average of sales is: {average} millions dollars')

df = df.sort_values('Sales', ascending=0)
plt.subplots(figsize=(6,8))
xaver = [df['Company'][0], 'Average of sales of other companies']
yaver = [df['Sales'][0], average]
plt.bar(xaver,yaver)
plt.title('Average of sales', fontsize='25', fontweight='bold')
plt.ylabel('Millions of dollars', fontweight='bold')
plt.show 
```

***Excluding walmart , the average of sales is 46602 millions of dollar***



-------------------

We need to exclude walmart in view of the fact that is a uncommon value, it can alter our judgment. As we can appreciate the sales of companies are extremely unequal, the average of sales is 14 times the sales of Walmart, it isn't determined for years working or number of stores, below we will see it.

**Facts**

* Even excluding Walmart, the difference of sales is strongly unequal between companies 

* Smaller retailers are so far from the top 5 



### 2. How much money did most companies sell?

```
plt.subplots(figsize=(6,10))
df = df.sort_values('Sales', ascending=0)
sns.barplot(x=df['Sales'], y = df['Company'])
#sns.despine(left=True, bottom=True)
plt.title('Money generated by company', fontsize='25', fontweight='bold')
plt.xlabel('Sales', fontweight='bold')
plt.ylabel('Companies', fontweight='bold')
```

***Walmart is the dominant company***

-------

The closer competitor in this case, is 5 times smaller than Wallmart. 


### 3. How many stores do companies have?

```
physical = df[(df['Stores'] > 0)]

df = df.sort_values('Stores', ascending=0)

plt.subplots(figsize=(6,10))
sns.barplot(x = physical['Stores'], y = df['Company'])
plt.title('Number of stores by company', fontsize='25',fontweight='bold')
plt.xlabel('number of stores', fontweight='bold')
plt.ylabel('Companies', fontweight='bold')
df['Stores'].mean()
```

* The number os stores is unequal, in this case **Dollar tree** is the company that has more stores, in contrast to their low sales

* The average number of stores is 3466, the most of companies are below

>Is the number of stores an important element for sell more?

Definitely, the number of stores is **not** the key to make more sales, the difference might be categories, marketing, strategies...

### 4. Is there any relation between number of stores and sales?

```
ax, g = plt.subplots(figsize=(20,9))

physical = physical.sort_values('Stores', ascending=0)
sns.set_theme(style='whitegrid')
sns.barplot(data = physical, x = 'Sales', y = 'Company', label='Sales', color='b')
sns.barplot(data = physical, x = 'Stores', y = 'Company', label ='Number of stores', color='red')
plt.xlabel('Sales', fontweight='bold')
plt.ylabel('Companies', fontweight='bold')
plt.title('Sales vs number of stores', fontweight='bold', fontsize='25')

ax.legend(ncol=2, loc='upper center', frameon =True)
```

**Walmart** is the company that make more sales in relation to number of stores

* **Kroger**, the more close competitor of Walmart in sales, has less stores. **Dollar Tree** has more stores than Walmart but is so far in sales. These companies are in **different categories**

### 5. What is the range of sales?

```
sns.boxplot(data=df, x='Sales', width=0.5, fliersize=10)
sns.despine(offset=20, trim=True)
plt.title('Range of sales', fontsize='25',fontweight='bold')
plt.xlabel('Sales', fontweight='bold')
```

The range of sales is between 18k and 65k

* In this graphic **Walmart** is clearly an outlier

### 6. What are the 5 companies that have more stores?

```
plt.subplots(figsize=(12,7))
physical = physical.sort_values('Stores', ascending=0)
sns.barplot(data=physical, x='Stores', y=physical['Company'][:5], palette='viridis')
plt.title('Top 5 companies that have more stores', fontsize='25',fontweight='bold')
plt.xlabel('Number of stores', fontweight='bold')
plt.ylabel('Companies', fontweight='bold')
```

 The companies that have more stores are **Dollar Tree, Dollar general, CSV incl. Targetm Walgreen boots and Walmart**.

 * The sales and categories are different between these companies, seeing that, we can affirm there isn't relation to sales

 * **Dollar Tree** have 3 times more stores than Walmart, and 44 times more than **HEB**, the company with the least number of stores


### 7. Which is the category of stores that sell more?

```
categories = df.groupby('Category').sum()

plt.subplots(figsize=(8,8))
sns.scatterplot(data=categories, x='Sales', y='Category', size='Sales', hue='Category', sizes=(100, 5000), legend=False)
plt.xlabel('Sales', fontweight='bold')
plt.ylabel('Categories of stores', fontweight='bold')
plt.title('Sales by categories', fontsize='25', fontweight='bold')
```

The categories that sell more are **Supercenters**, followed by **Grocery** and **Drug stores**

* Curiosly, Walmart is the principal Supercenter, the other companies in this category have low sales
